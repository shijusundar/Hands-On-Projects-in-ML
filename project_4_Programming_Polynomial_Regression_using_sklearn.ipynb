***Programming Polynomial Regression using sklearn***

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
x=np.linspace(0,0.5,100)
x.shape
```

```python
y_actual=np.cos(2*np.pi*x)**2
y_actual.shape
```

```python
plt.plot(x,y_actual)
```

```python
noise=np.random.normal(0,0.1,100)
```

```python
y=y_actual+noise
```

```python
plt.plot(x,y_actual)
plt.scatter(x,y,color='red')
```

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
print(x.shape,y.shape)
```

```python
model=linear_model.LinearRegression(fit_intercept=True)
model.fit(x,y)
```

```python
y_pred=model.predict(x)
```

```python
plt.plot(x,y_actual)
plt.scatter(x,y,color='red')
plt.plot(x,y_pred,color='green')
```

```python
deg=2
scalar=StandardScaler()
polynomial_features=PolynomialFeatures(degree=deg,include_bias=True)
linear=linear_model.LinearRegression(fit_intercept=True)
pipeline=Pipeline([("polynomial",polynomial_features),("scalling",scalar),("linear",linear)])
```

```python
pipeline.fit(x,y)
```

```python
y_pred=pipeline.predict(x)
```

```python
plt.plot(x,y_actual)
plt.scatter(x,y,color='red')
plt.plot(x,y_pred,color='green')
```

```python
deg=1000
scalar=StandardScaler()
polynomial_features=PolynomialFeatures(degree=deg,include_bias=True)
linear=linear_model.LinearRegression(fit_intercept=True)
pipeline=Pipeline([("polynomial",polynomial_features),("scalling",scalar),("linear",linear)])
pipeline.fit(x,y)
y_pred=pipeline.predict(x)
plt.plot(x,y_actual)
plt.scatter(x,y,color='red')
plt.plot(x,y_pred,color='green')
```
